# Práctica de Python: If y Else

En esta práctica aprenderás a usar las estructuras condicionales **if** y **else** en Python.

Recuerda:
```python
if condición:
    # código si la condición es verdadera
else:
    # código si la condición es falsa
```

## Ejercicio 1
Pide al usuario que ingrese su edad y muestra:
- 'Eres mayor de edad' si tiene 18 años o más.
- 'Eres menor de edad' si tiene menos de 18.

Escribe tu código en la celda siguiente:

````python
v1=int(input("dame un numero"))
if v1>=18:
  print("eres mayor de edad")
else:
        print("eres menor de edad")
````

## Ejercicio 2
Pide al usuario que ingrese un número. Si el número es **positivo** muestra 'Es positivo', si es **negativo** muestra 'Es negativo', si es **cero** muestra 'Es cero'.

Escribe tu código en la celda siguiente:

````python
v1=int(input("dame un numero:"))
if v1>0:
  print("es positivo")
elif v1<0:
    print("es negativo")
else:
      print("es cero")
````

## Ejercicio 3
Pide al usuario que ingrese una contraseña. Si la contraseña es `'python123'` muestra 'Acceso permitido'. Si no, muestra 'Contraseña incorrecta'.

Escribe tu código en la celda siguiente:

````python
v1=input("dame una contraseña")
if v1=="python123":
  print("acceso permitido")
else:
    print("contraseña incorrecta")
````

## Ejercicio 4
Pide al usuario que ingrese su calificación (0 a 100). Si la calificación es **mayor o igual a 60**, imprime 'Aprobado'. De lo contrario, imprime 'Reprobado'.

Escribe tu código en la celda siguiente:

````python
v1=int(input("dame una calificacion:"))
if v1>=60:
  print("aprobado")
elif v1<60:
    print("reprobado")
````
